```python
import os
for i in os.listdir('final_data/train/labels'):
    with open('final_data/train/labels/'+i,'r') as f:
        lines=f.readlines()
    new_lines = []
    for line in lines:
        stripped = line.lstrip()  # remove leading spaces
        if stripped.startswith("2"):
            parts = stripped.split()
            parts[0] = "0"  # replace the first number
            new_line = " ".join(parts) + "\n"
            new_lines.append(new_line)
    with open('final_data/train/labels/'+i, "w") as f:
        f.writelines(new_lines)  
```

```python
import os
import glob
path = '/Users/manasganti/Fall-2025/comp_vision/soccer-player-tracking/NTERN-2'
fin_imgs = []
fin_labs = []
for i in os.listdir(path):
    if i == 'test' or i == 'train' or i == 'valid':
        p = os.path.join(path,i,'labels')
        for j in os.listdir(p):
            with open(os.path.join(p,j) , 'r') as f:
                lines = f.readlines()
            cleaned_lines = []
            for line in lines:
                if line.strip() == "":
                    continue
                first = line.split()[0]  # get first value
                if first in ["0", "1", "4"]:
                    continue  # skip lines starting with 0, 1, or 4
                elif first in ["2", "3"]:
                    # replace leading 2 or 3 with 0
                    parts = line.split()
                    parts[0] = "0"
                    cleaned_lines.append(" ".join(parts) + "\n")
            with open(os.path.join(p,j) , 'w') as f:
                f.writelines(cleaned_lines)
```
