```python
import os
import shutil
import random

def split_dataset(image_paths, label_paths, output_dir, train_ratio=0.8, seed=42):
    """
    Splits a dataset of images and labels into train/val folders.

    Parameters:
        image_paths (list[str]): List of paths to images.
        label_paths (list[str]): List of paths to corresponding labels.
        output_dir (str): Directory to save the train/val split.
        train_ratio (float): Ratio of training data (default 0.8).
        seed (int): Random seed for reproducibility.
    """
    print(len(image_paths) , len(label_paths))
    assert len(image_paths) == len(label_paths), "Images and labels must be paired equally."

    # Shuffle dataset
    data = list(zip(image_paths, label_paths))
    random.seed(seed)
    random.shuffle(data)

    # Split
    split_index = int(len(data) * train_ratio)
    train_data = data[:split_index]
    val_data = data[split_index:]

    # Create output directories
    for split in ["train", "val"]:
        os.makedirs(os.path.join(output_dir, split, "images"), exist_ok=True)
        os.makedirs(os.path.join(output_dir, split, "labels"), exist_ok=True)

    # Copy files
    for split_name, split_data in [("train", train_data), ("val", val_data)]:
        for img_path, lbl_path in split_data:

            shutil.copy(os.path.join('/Users/manasganti/Fall-2025/comp_vision/soccer-player-tracking/final_data/images',img_path), os.path.join(output_dir, split_name, "images", os.path.basename(img_path)))
            shutil.copy(os.path.join('/Users/manasganti/Fall-2025/comp_vision/soccer-player-tracking/final_data/labels',lbl_path), os.path.join(output_dir, split_name, "labels", os.path.basename(lbl_path)))

    print(f"✅ Split complete: {len(train_data)} train, {len(val_data)} val samples.")
    print(f"Saved under: {output_dir}")


# Example usage
if __name__ == "__main__":
    # Example lists of paths
    images = os.listdir('/Users/manasganti/Fall-2025/comp_vision/soccer-player-tracking/final_data/images')
    labels = os.listdir('/Users/manasganti/Fall-2025/comp_vision/soccer-player-tracking/final_data/labels')

    split_dataset(images, labels, output_dir="/Users/manasganti/Fall-2025/comp_vision/soccer-player-tracking/final_data", train_ratio=0.8)
```
